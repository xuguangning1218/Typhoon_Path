```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
Multi_ConvGRU_i_1 = [3.74, 3.75,3.76]
Multi_ConvGRU_i_2 = [6.07, 6.08,6.09]
Multi_ConvGRU_i_3 = [8.39, 8.40,8.41]
Multi_ConvGRU_i_4 = [10.71, 10.72,10.73]
Multi_ConvGRU_i_5 = [13.03, 13.04,13.05]
```

```python
X = np.array([0,1,2])
gap = 0.18
width = 0.18
```

```python
plt.figure(figsize=(8,6))
plt.bar(X+gap*(-2), Multi_ConvGRU_i_1, width=width, label= 'Multi-ConvGRU(i=1)')
plt.bar(X+gap*(-1), Multi_ConvGRU_i_2, width=width, label= 'Multi-ConvGRU(i=2)')
plt.bar(X+gap*0, Multi_ConvGRU_i_3, width=width, label= 'Multi-ConvGRU(i=3)')
plt.bar(X+gap*1, Multi_ConvGRU_i_4, width=width, label= 'Multi-ConvGRU(i=4)')
plt.bar(X+gap*2, Multi_ConvGRU_i_5, width=width, label= 'Multi-ConvGRU(i=5)')

for a,b in zip(X+gap*(-2),Multi_ConvGRU_i_1):
    plt.text(a, b+0.05, '%.2f' % b, ha='center',fontsize=9)
for a,b in zip(X+gap*(-1),Multi_ConvGRU_i_2):
    plt.text(a, b+0.05, '%.2f' % b, ha='center',fontsize=9)
for a,b in zip(X+gap*(0),Multi_ConvGRU_i_3):
    plt.text(a, b+0.05, '%.2f' % b, ha='center',fontsize=9)
for a,b in zip(X+gap*(1),Multi_ConvGRU_i_4):
    plt.text(a, b+0.05, '%.2f' % b, ha='center',fontsize=9)
for a,b in zip(X+gap*(2),Multi_ConvGRU_i_5):
    plt.text(a, b+0.05, '%.2f' % b, ha='center',fontsize=9)
    
plt.xticks(X,[4,8,16],fontsize=15)
plt.yticks(fontsize=15)
plt.ylabel('Numbers of Parameters (M)', fontsize=15)
plt.xlabel('Numbers of Isobaric Planes', fontsize=15)
plt.legend(loc='center', bbox_to_anchor=(0.5, 1.08),ncol=3)
plt.savefig('parameter.pdf',bbox_inches='tight')
plt.show()
```
